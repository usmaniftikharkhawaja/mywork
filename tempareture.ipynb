```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv(r'C:\Users\Usman-PC\Documents\Temperature_Pakistan\Tempreture_1901_2016_Pakistan.csv')

# Extract the necessary columns
temperature = data['Temperature - (Celsius)']
year = data[' Year']  # Notice the space before 'Year'
month = data['Month']

# Sample the data for better readability
sample_indices = range(0, len(year), 12)  # Select one data point per year
sample_temperature = temperature[sample_indices]

# Bar Graph
plt.figure(figsize=(12, 6))
plt.bar(year[sample_indices], sample_temperature, color='b')
plt.xlabel('Year')
plt.ylabel('Temperature (Celsius)')
plt.title('Temperature Changes in Pakistan (1901-2016) - Bar Graph')
plt.grid(True)
plt.show()

# Histogram
plt.figure(figsize=(12, 6))
plt.hist(sample_temperature, bins=20, color='b')
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Frequency')
plt.title('Temperature Distribution in Pakistan (1901-2016) - Histogram')
plt.grid(True)
plt.show()
```
